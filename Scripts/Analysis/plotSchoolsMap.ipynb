```python
columns = ['Nome',
           'Categoria Administrativa',
           'Latitude',
           'Longitude']

df_schools = pd.DataFrame(columns=columns)
df_schools
```

```python
school = {}

for document in collection.find():
    if (document['Latitude']):
            school = {'Nome': document['Escola'], 
                      'Latitude': document['Latitude'], 
                      'Longitude': document['Longitude'],
                      'Categoria Administrativa': document['Categoria Administrativa']}
            df_schools = df_schools.append(school, ignore_index=True)
```

### Mapa de Pelotas

```python
def mapPelotas():
    return folium.Map(location=[-31.7154, -52.3376],
                      zoom_start=12.3,)
```

**Escolas no Mapa**

```python
mapSchools = mapPelotas();

for index, row in df_schools.iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup='<strong>ESCOLA</strong>:' + row['Nome'],
        icon=folium.Icon(markerColor(row['Categoria Administrativa'])), 
        tooltip='Click aqui!'
        ).add_to(mapSchools)
```

```python
folium.TileLayer('Stamen Terrain').add_to(mapSchools)
folium.TileLayer('Stamen Toner').add_to(mapSchools)
folium.TileLayer('Stamen Water Color').add_to(mapSchools)
folium.TileLayer('cartodbpositron').add_to(mapSchools)
folium.TileLayer('cartodbdark_matter').add_to(mapSchools)
folium.LayerControl().add_to(mapSchools)
```

**Mapa de Calor**

```python
heatMapSchools = mapPelotas();

hm_wide = HeatMap(list(zip(df_schools.Latitude.values, df_schools.Longitude.values)),
                     min_opacity=0.2,
                     radius=17, blur=15,
                     max_zoom=1 
                 )
```

```python
heatMapSchools.add_child(hm_wide)
```
